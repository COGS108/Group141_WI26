## Research Question

To what extent is a county's median household income associated with its voter turnout in the 2024 U.S. Presidential Election, and does this relationship persist when controlling for educational attainment and urbanization?

-Voter turnout will be defined as the total votes cast for president divided by the Citizen Voting Age Population (CVAP). We will utilize Pearson/Spearman correlation coefficients and multiple linear regression to estimate the change in turnout percentage points per $10,000 increase in median household income (using ACS 5-year estimate B19013). This study aims to determine if income remains a statistically significant predictor of civic engagement at the aggregate county level after accounting for demographic confounders.

## Background and Prior Work


<p>
<a id="ref1"></a>
Voter turnout is a core measure of democratic participation, but participation in elections is highly unequal across socioeconomic groups in the United States: richer people are more likely to vote than poorer people.<sup><a href="#fn1">1</a></sup>
Recent U.S. elections have also seen unusually high turnout by historical standards. For example, Pew Research Center reports that 2020 turnout reached levels not seen in decades, estimating that 62.8% of the voting-age population voted, while noting that U.S. turnout still trails many peer democracies under common cross-national turnout measures.<sup><a href="#fn2">2</a></sup>
</p>

<p>
<a id="ref2"></a>
Income-based inequalities in participation are important because they can affect political representation. When lower-income citizens vote at lower rates, their interests and preferences are less likely to be reflected in public policy, even if income itself is not necessarily the direct cause of voting behavior.<sup><a href="#fn1">1</a></sup>
At the same time, concerns about unequal participation are heightened in a period of rising economic inequality in the United States; Inequality.org summarizes that the top 1% share of national income has increased substantially since 1970 (citing the World Inequality Database).<sup><a href="#fn3">3</a></sup>
</p>

<p>
<a id="ref3"></a>
Prior research has examined how income inequality relates to turnout and to “income bias” in turnout (the degree to which higher-income people vote more than lower-income people), but has reached mixed conclusions.
Matsubayashi and Sakaiya (2021), using large-scale cross-national survey data, argue that higher income inequality can be associated with lower income bias in turnout by demobilizing higher-income citizens while mobilizing lower-income citizens, and they discuss potential mechanisms such as vote buying.<sup><a href="#fn4">4</a></sup>
At the same time, more recent precinct-level work emphasizes that a major obstacle in this literature is measurement: survey-based turnout can be affected by response and social desirability bias, which can produce widely varying estimates of income-based turnout gaps across studies.<sup><a href="#fn5">5</a></sup>
</p>

<p>
<a id="ref4"></a>
Because voting behavior is shaped by multiple factors that correlate with income (e.g., education, age composition, urbanization), our project examines whether areas with higher median household incomes systematically exhibit different turnout levels and whether this association remains after accounting for key demographic and structural characteristics.<sup><a href="#fn1">1</a></sup><sup><a href="#fn2">2</a></sup>
</p>

## Hypothesis


We hypothesize that U.S. counties with higher median household income will have higher voter turnout in the 2024 presidential election, defined as total presidential votes divided by CVAP. After controlling for educational attainment and urbanization, we expect the income–turnout association to attenuate but remain positive because income is linked to resources and lower barriers to political participation beyond education and place-based differences.

## Data

Unit of Analysis:
The analysis unit for this project is county. It includes all 58 California counties.

Turnout Data:
Voter turnout data for November 5, 2024 by county was obtained from the California Secretary of State. We measure voter turnout in two different ways. The first measure is turnout among registered voters, calculated as the number of ballots cast divided by the number of registered voters in each county. The second measure is turnout among the eligible population, calculated as the number of ballots cast divided by the number of eligible voters. Both measurement methods help the analysis determine whether the results vary depending on how voter turnout is defined.

Income Data:
The main independent variable is median household income at the county level. This data comes from the U.S. Census Bureau’s American Community Survey (ACS) 5-year estimates. We use the county-level median household income variable because it is consistently measured across counties and commonly used in economic and social research.

Control Variables:
To consider other factors related to voter turnout, we plan to include a small number of control variables at the county level. This may include educational background (adult ratio above bachelor's degree), age composition (resident ratio above 65), and population density as a simple indicator of urbanization. These variables can also be seen in public census data.

Data Size and Format:
The final dataset contains one line for each of California's 58 counties. Data is saved as a CSV file and combined using a county identifier. Also, we will check for missing or mismatched values before the analysis.

Data Limitations:
There are some limitations to the data. ACS income is estimated for the year before the 2024 election, and voter turnout statistics vary depending on the definition used by the state government. For this reason, the results are interpreted as county-level relevance rather than causal influence.
